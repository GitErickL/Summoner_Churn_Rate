# Getting Data for Large Amount of Players

```python
watcher = LolWatcher(api_key = "RGAPI-f922ea08-3c79-4dc6-b807-adbb20db5bb3")
puddingpiler = watcher.summoner.by_name(region = 'na1', summoner_name = "puddingpiler")
```

We ran a previous script that fetched a ton of summoners' puuids and saved them to a file. Now we unpack these puuids to a list in order to get their game data.

```python
with open("puuid_list.txt", "r") as f:
    puuids = f.read()
    

puuids = ast.literal_eval(puuids)

puuids = [n.strip() for n in puuids]
```

```python
now = 1653528536
add_4_days = 60*60*24*4

for i, puuid in enumerate(puuids):
    
        
    match_list = []


    add_time = 0 

    print("getting matches of player {}".format(i))
    while beginning + add_time < now:

        matches = watcher.match.matchlist_by_puuid(
                                                     region = 'americas', 
                                                     puuid = puuid,
                                                     #queue = 420,
                                                     start_time = beginning + add_time,
                                                     end_time = beginning + add_4_days + add_time,
                                                     count = 100
                                                    )
        add_time += add_4_days

        if len(matches) == 0:
            continue

        else:
            for match in matches:
                match_list.append(match)

    


    for j, match in enumerate(match_list):
        
        
        print("getting data for match {} of {}".format(j, len(match_list)))

        while True:
            
            try:
                
                match_json = watcher.match.by_id(region = 'americas', match_id =  match)
                start_time = match_json["info"]["gameStartTimestamp"]
                duration   = match_json["info"]["gameDuration"]
                
                participants = match_json["info"]["participants"]
                
                for i in range(len(participants)):
                
                    if participants[i]["puuid"] == puuid:        
                
                        win = participants[i]["win"]
                                 

            except HTTPError:
                
                print("Got HTTPError, trying again.")
                continue

                
            break



        with open("Player_Time_Data_Large/player_{}_times.txt".format(i), "a") as f:

            f.write(puuid + " " + match + " " + str(start_time) + " " + str(duration) + " " + str(win) + "\n")
            

        time.sleep(2)
        
        
```
