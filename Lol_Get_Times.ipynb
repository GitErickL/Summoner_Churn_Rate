```python
with open("puuid_list.txt", "r") as f:
    puuids = f.read()
    
import ast
puuids = ast.literal_eval(puuids)

puuids = [n.strip() for n in puuids]
```

For each player
    Get list of match_id's for that player since the beginning of available time
        From each of these match_id's get the start times of each game

```python
now = 1653528536
add_4_days = 60*60*24*4
```

```python
for i, puuid in enumerate(puuids):
    if i < 3:
        continue
        
    match_list = []


    add_time = 0 

    print("getting matches")
    while beginning + add_time < now:

        matches = watcher.match.matchlist_by_puuid(
                                                     region = 'americas', 
                                                     puuid = puuid,
                                                     #queue = 420,
                                                     start_time = beginning + add_time,
                                                     end_time = beginning + add_4_days + add_time,
                                                     count = 100
                                                    )
        add_time += add_4_days

        if len(matches) == 0:
            continue

        else:
            for match in matches:
                match_list.append(match)

    


    for j, match in enumerate(match_list):
        print("getting times for match {} of {}".format(j,len(match_list)))

        while True:
            try:
                start_time = watcher.match.by_id(region = 'americas', match_id =  match)["info"]["gameStartTimestamp"]
                duration   = watcher.match.by_id(region = 'americas', match_id =  match)["info"]["gameDuration"]


            except HTTPError as e:
                print("got HTTPError. Trying again")
                continue

            break



        with open("Player_Time_Data/player_{}_times.txt".format(i),"a") as f:

            f.write(puuid + " " + match + " " + str(start_time) + " " + str(duration) + "\n")
            #f.write(str(time_start_list[i]) + " " + str(time_end_list[i]) + "\n")

        time.sleep(2)
```
